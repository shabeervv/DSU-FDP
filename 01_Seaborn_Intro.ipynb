## More Examples

\#3. Seaborn allows for additional plot types and easy plotting functionality

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

cyl_4 = cars[cars["cylinders"] == 4]
cyl_8 = cars[cars["cylinders"] == 8]

plt.figure(figsize=(8,6))

sns.kdeplot(
    data=cyl_4,
    x="horsepower",
    y="mpg",
    fill=True,
    cmap="Blues"
)
#kernel density estimate (KDE)
sns.kdeplot(
    data=cyl_8,
    x="horsepower",
    y="mpg",
    fill=True,
    cmap="Reds"
)

plt.xlabel("Horsepower")
plt.ylabel("Miles per Gallon (MPG)")
#To add label
plt.annotate("4 Cylinders", (105, 32), color="blue",
             fontsize=16, fontweight="bold")

plt.annotate("8 Cylinders", (190, 18), color="red",
             fontsize=16, fontweight="bold")

plt.tight_layout()
#Automatically adjusts spacing so labels and the graph fit neatly.
plt.show()
```

```python
sns.pairplot(cars[cars.cylinders.isin([4,6,8])],
             hue='cylinders',
             vars=['mpg', 'horsepower', 'acceleration'],
             #palette='inferno_r'
             palette="plasma"
            );
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
cars = sns.load_dataset('mpg')

# Filter only USA cars
usa = cars[cars['origin'] == 'usa']

# Remove rows with missing horsepower values
usa = usa.dropna(subset=['horsepower'])

# Create the plot
plt.figure(figsize=(8, 6))

sns.violinplot(
    x='cylinders',
    y='horsepower',
    data=usa,
    scale='width',
    inner=None,
    color='forestgreen'
)

sns.swarmplot(
    x='cylinders',
    y='horsepower',
    data=usa,
    color='white',
    size=3
)

plt.title('Horsepower Distribution of USA Cars by Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Horsepower')
plt.show()
```

```python
sns.set_style('white')
pts = [1, 1.5, (7/3), 3, 3.75, 4.25, 5, 19/3, 8, 12]

plt.figure(figsize=(6,4))
sns.boxplot(pts, color='xkcd:golden yellow', width=0.4)
#sns.despine(left=True)
#Removes the left border (spine) of the plot.
plt.xlim(-3, 16)
plt.xticks(range(0,13,2))
#Displays tick marks at:0  2  4  6  8  10  12
plt.tight_layout();
#Automatically adjusts spacing so labels and plot elements don't overlap.
```

Histogram
KDE (Kernel Density Estimate) curve

```python
plt.figure(figsize=(8,5))
#Distribution Plot
sns.distplot(cars.weight,
             bins=20,
             kde_kws={"lw": 8, 'color':'xkcd:bluish green'},
             hist_kws={'alpha': 0.3, 'color':'gray'}
            )

sns.despine(left=True)
plt.xlabel('Weight (lbs)', fontsize=16, labelpad=15)
plt.yticks([])
plt.tight_layout();
```

```python
plt.figure(figsize=(8,6))
#categorical swarm plot
g = sns.catplot(x='origin', y='horsepower', hue='cylinders',
                kind='swarm',
                palette=sns.color_palette('inferno_r', 3,),
                data=cars[cars.cylinders.isin([4, 6, 8])]
               )

plt.xlabel('\nOrigin')
plt.ylabel('Horsepower')
plt.gca().set_xticklabels(['USA', 'Japan', 'Europe'])
g._legend.set_title('Cylinders')
plt.tight_layout();
```
