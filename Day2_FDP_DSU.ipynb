# **Nonlocal and Namespaces**

```python
#Program 4: Nonlocal Variable
def outer():
    number = 10
    def inner():
        nonlocal number
        number += 5
        print(number)
    inner()
outer()
```

//allows an inner function to access and modify a variable from its enclosing (outer) function.

```python
#Program 5: globals()
name = "Python"

print(globals()["name"])
```

globals() returns a dictionary containing all global variables and functions in the program.

```python
name = "Raj"
age = 25
city = "Chennai"

def update():
    globals()["name"] = "Rahul"
    globals()["age"] += 1
    globals()["city"] = "Bengaluru"
    print(globals())
update()

print(name)
print(age)
print(city)
print(globals())
```

```python
#Program 6: locals()
def student():

    name = "John"
    age = 20

    print(locals())

student()
```

# **Classes and Objects**

Class

Explanation: A class is a blueprint for creating objects. It defines the attributes (data) and methods (functions).
Key: Template for objects.

Object

Explanation: An object is an instance of a class. It uses the properties and methods defined in the class.
Key: Real-world instance of a class

```python
#Program 7: Student Class
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)
s1 = Student("Raj", 90)
s1.display()
```

```python
#Program 8: Rectangle Class
class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
    def area(self):
        return self.length * self.breadth
r = Rectangle(10,5)
print("Area =", r.area())
```

**Inheritance**

```python
#Program 10: Single Inheritance
class Animal:
    def sound(self):
        print("Animal Sound")
class Dog(Animal):
    def bark(self):
        print("Barking")
d = Dog()
d.sound()
d.bark()
```

```python
#Program 11: Multilevel Inheritance
class A:
    def show(self):
        print("Class A")
class B(A):
    pass
class C(B):
    pass
obj = C()
obj.show()
```

```python
#Program 12: Method Overriding
class Bird:
    def fly(self):
        print("Bird Flies")
class Sparrow(Bird):
    def fly(self):
        print("Sparrow Flies")
s = Sparrow()
s.fly()
```

**User-Defined Exception & Mini Project**

```python
#Program 16: User-Defined Exception
class InvalidAge(Exception):
    pass
try:
    age = int(input("Age: "))
    if age < 18:
        raise InvalidAge("Not Eligible")
    print("Eligible")
except InvalidAge as e:
    print(e)
```

```python
#Program 17: ATM Withdrawal
class LowBalance(Exception):
    pass

balance = 5000

try:

    amount = int(input("Withdraw Amount: "))

    if amount > balance:
        raise LowBalance("Insufficient Balance")

    balance -= amount

    print("Remaining Balance:", balance)

except LowBalance as e:

    print(e)
```

```python
#Program 18: Student Result System (Mini Project)
class Student:

    def __init__(self, name, marks):

        self.name = name
        self.marks = marks

    def result(self):

        if self.marks >= 50:
            print("Pass")
        else:
            print("Fail")

s = Student("Raj", 65)

print("Student:", s.name)

s.result()
```
