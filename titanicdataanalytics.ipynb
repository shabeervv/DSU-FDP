```python
# ==========================================
# DATA ANALYTICS USING PYTHON
# Titanic Dataset Analysis
# ==========================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Load Dataset
# -----------------------------
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

# -----------------------------
# Display Dataset
# -----------------------------
print("First 5 Records")
print(df.head())

print("\nDataset Shape")
print(df.shape)

print("\nColumn Names")
print(df.columns)

# -----------------------------
# Basic Information
# -----------------------------
print("\nDataset Information")
print(df.info())

print("\nSummary Statistics")
print(df.describe())

# -----------------------------
# Missing Values
# -----------------------------
print("\nMissing Values")
print(df.isnull().sum())

# Fill missing Age with Median
df["Age"].fillna(df["Age"].median(), inplace=True)

# Fill missing Embarked with Mode
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)

# Drop Cabin column
df.drop("Cabin", axis=1, inplace=True)

print("\nMissing Values After Cleaning")
print(df.isnull().sum())

# -----------------------------
# Survival Analysis
# -----------------------------
survival = df["Survived"].value_counts()

print("\nSurvival Count")
print(survival)

plt.figure(figsize=(6,4))
survival.plot(kind='bar')
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Passengers")
plt.show()

# -----------------------------
# Gender Analysis
# -----------------------------
gender = df["Sex"].value_counts()

plt.figure(figsize=(6,4))
gender.plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# -----------------------------
# Passenger Class Distribution
# -----------------------------
pclass = df["Pclass"].value_counts()

plt.figure(figsize=(6,4))
pclass.plot(kind='bar')
plt.title("Passenger Class")
plt.xlabel("Class")
plt.ylabel("Passengers")
plt.show()

# -----------------------------
# Age Distribution
# -----------------------------
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# -----------------------------
# Fare Distribution
# -----------------------------
plt.figure(figsize=(8,5))
plt.hist(df["Fare"], bins=20)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Passengers")
plt.show()

# -----------------------------
# Correlation
# -----------------------------
numeric_df = df.select_dtypes(include=np.number)

correlation = numeric_df.corr()

print("\nCorrelation Matrix")
print(correlation)

plt.figure(figsize=(8,6))
plt.imshow(correlation, cmap='coolwarm')
plt.colorbar()

plt.xticks(range(len(correlation.columns)),
           correlation.columns,
           rotation=90)

plt.yticks(range(len(correlation.columns)),
           correlation.columns)

plt.title("Correlation Matrix")
plt.show()

# -----------------------------
# Group Analysis
# -----------------------------
print("\nAverage Fare by Passenger Class")
print(df.groupby("Pclass")["Fare"].mean())

print("\nAverage Age by Gender")
print(df.groupby("Sex")["Age"].mean())

print("\nSurvival Rate by Gender")
print(df.groupby("Sex")["Survived"].mean())

print("\nSurvival Rate by Passenger Class")
print(df.groupby("Pclass")["Survived"].mean())

# -----------------------------
# Top 10 Highest Fare
# -----------------------------
print("\nTop 10 Highest Fare Passengers")
print(df.nlargest(10, "Fare")[["Name","Fare","Pclass"]])

# -----------------------------
# Save Cleaned Dataset
# -----------------------------
df.to_csv("cleaned_titanic.csv", index=False)

print("\nCleaned dataset saved successfully.")
```

```python
#Numpy
import numpy as np

# Original Array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Original Array:")
print(arr)

# empty_like
print("\n1. empty_like()")
print(np.empty_like(arr))

# ones_like
print("\n2. ones_like()")
print(np.ones_like(arr))

# zeros_like
print("\n3. zeros_like()")
print(np.zeros_like(arr))

# full_like
print("\n4. full_like(value=9)")
print(np.full_like(arr, 9))

# empty
print("\n5. empty(shape=(2,3))")
print(np.empty((2, 3)))

# ones
print("\n6. ones(shape=(2,3))")
print(np.ones((2, 3)))

# zeros
print("\n7. zeros(shape=(2,3))")
print(np.zeros((2, 3)))

# full
print("\n8. full(shape=(2,3), value=5)")
print(np.full((2, 3), 5))

# copy
print("\n9. copy()")
copy_arr = arr.copy()
copy_arr[0][0] = 100

print("Original Array:")
print(arr)

print("Copied Array:")
print(copy_arr)
```
